```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
# https://stepik.org/media/attachments/course/4852/training_mush.csv
df = pd.read_csv('training_mush.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
clf = RandomForestClassifier(random_state=0)
clf
```

```python
params = {'n_estimators': range(10,50,10),
         'max_depth': range(1,12,2),
         'min_samples_leaf': range(1,7),
         'min_samples_split': range(2,9,2)}
```

```python
gs = sklearn.model_selection.GridSearchCV(clf, params, cv=3, n_jobs=-1)
```

```python
x = df.drop('class', axis=1)
y = df['class']
```

```python
gs.fit(x, y)
```

```python
gs.best_estimator_
```

```python
d = pd.DataFrame({'name': x.columns,
             'imp': gs.best_estimator_.feature_importances_})
d.sort_values(by=['imp'], ascending=False)
```

```python
# https://stepik.org/media/attachments/course/4852/testing_mush.csv
df2 = pd.read_csv('testing_mush.csv')
df2.head()
```

```python
p = gs.predict(df2)
p.sum()
```
