```python
# https://stepik.org/media/attachments/course/4852/iris.csv
df = pd.read_csv('iris.csv', index_col=0);
df.head(3)
```

```python
sns.kdeplot(df);
```

```python
sns.pairplot(df);
```

```python
sns.distplot(df['petal length'], color='blue');
sns.distplot(df['petal width'], color='red');
sns.distplot(df['sepal width'], color='yellow');
sns.distplot(df['sepal length'], color='black');
```

```python
for column in df:
    #print(column)
    sns.kdeplot(df[str(column)])
```

```python
colors = ['g', 'r', 'blue', 'yellow', 'white']
for col, color in zip(df.iloc[:, :-1], colors):
    sns.distplot(df[col], kde_kws = {'color':color, 'lw':1, 'label':col})
```

```python
sns.pairplot(df,  vars=df.columns[:-1], diag_kind='kde')
```

```python
df.iloc[:, 0:6].hist();
```

```python
df.iloc[:,0:4].apply(lambda x: sns.kdeplot(x))
```

```python
for c in df.drop('species', axis=1):
    sns.kdeplot(df[c], shade=True)
```

```python
for i in df.iloc[:, 0:-1]:
    df[str(i)].plot(kind='kde',legend=True)
```
